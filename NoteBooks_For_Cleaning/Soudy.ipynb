## CAMPAIGN_TOUCH & CATEGORY & CHANNEL

```python
import pandas as pd

# Import CAMPAIGN_TOUCH, CATEGORY and CHANNEL files
campaign_touch = pd.read_csv('../DataSet/CAMPAIGN_TOUCH.csv')
category = pd.read_csv('../DataSet/CATEGORY.csv')
channel = pd.read_csv('../DataSet/CHANNEL.csv')




   #----Cleaning----#
# ---1--- convert touch_date to date datatype
campaign_touch['touch_date'] = pd.to_datetime(campaign_touch['touch_date'])

# ---2--- encode outcome 


campaign_touch['outcome'] = campaign_touch['outcome'].map({
    'Ignored': 0,
    'Clicked': 1,
    'Converted': 2
})



# Display first few rows
print("-------------------CAMPAIGN_TOUCH Dataset:-------------------")
print(campaign_touch.head())
print("\n-------------------CATEGORY Dataset:-------------------")
print(category.head())
print("\n-------------------CHANNEL Dataset:-------------------")
print(channel.head())



#Validation
```

# run this after finish all cleaning process

```python
# Save cleaned datasets
campaign_touch.to_csv('../Cleaned_DataSet/CAMPAIGN_TOUCH.csv', index=False)
category.to_csv('../Cleaned_DataSet/CATEGORY.csv', index=False)
channel.to_csv('../Cleaned_DataSet/CHANNEL.csv', index=False)
print("Cleaned datasets saved to Cleaned_DataSet folder")
```
