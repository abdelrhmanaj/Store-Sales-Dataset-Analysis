## CUSTOMER & MARKETING_CHANNEL & EXTERNAL_FACTOR

```python
import pandas as pd

# Import CUSTOMER, MARKETING_CHANNEL and EXTERNAL_FACTOR files
customer = pd.read_csv('../DataSet/CUSTOMER.csv')
marketing_channel = pd.read_csv('../DataSet/MARKETING_CHANNEL.csv')
external_factor = pd.read_csv('../DataSet/EXTERNAL_FACTOR.csv')

# Display first few rows
print("CUSTOMER Dataset:")
print(customer.head())
print("\nMARKETING_CHANNEL Dataset:")
print(marketing_channel.head())
print("\nEXTERNAL_FACTOR Dataset:")
print(external_factor.head())

```

```python
import pandas as pd
import numpy as np

# --- 1. Customer cleaning ---
customer = customer.copy()
# Convert signup_date to datetime
customer['signup_date'] = pd.to_datetime(customer['signup_date'], errors='coerce')
# Normalize city and state
customer['city'] = customer['city'].astype(str).str.strip().str.title()
customer['state'] = customer['state'].astype(str).str.strip().str.title()
# Handle missing values
customer['status'] = customer['status'].fillna('Unknown')
customer = customer.dropna(subset=['customer_id'])

# --- 2. Marketing Channel cleaning ---
marketing_channel = marketing_channel.copy()
# Normalize channel name and type
marketing_channel['channel_name'] = marketing_channel['channel_name'].astype(str).str.strip()
# Ensure cost_tier is consistent
marketing_channel['cost_tier'] = marketing_channel['cost_tier'].astype(str).str.strip().str.title()

# --- 3. External Factor cleaning ---
external_factor = external_factor.copy()
# Convert factor_date to datetime
external_factor['factor_date'] = pd.to_datetime(external_factor['factor_date'], errors='coerce')
# Ensure numeric columns are correct
num_cols = ['temp_c', 'rainfall_mm', 'trend_index', 'cpi_index']
for col in num_cols:
    external_factor[col] = pd.to_numeric(external_factor[col], errors='coerce')
# Handle nulls in external factors with median (أفضل للبيانات الجوية والاقتصادية)
external_factor[num_cols] = external_factor[num_cols].fillna(external_factor[num_cols].median())

# --- Validation ---
print("--- CUSTOMER Info ---")
print(customer.info())
print("\n--- EXTERNAL_FACTOR Summary ---")
print(external_factor.describe())
```

# run this after finish all cleaning process

```python
# Save cleaned datasets
customer.to_csv('../Cleaned_DataSet/CUSTOMER.csv', index=False)
marketing_channel.to_csv('../Cleaned_DataSet/MARKETING_CHANNEL.csv', index=False)
external_factor.to_csv('../Cleaned_DataSet/EXTERNAL_FACTOR.csv', index=False)
print("Cleaned datasets saved to Cleaned_DataSet folder")
```
