## PROMOTION & STORE & SUPPORT_TICKET

```python
import pandas as pd

# Import PROMOTION, STORE and SUPPORT_TICKET files
promotion = pd.read_csv('../DataSet/PROMOTION.csv')
store = pd.read_csv('../DataSet/STORE.csv')
support_ticket = pd.read_csv('../DataSet/SUPPORT_TICKET.csv')

# Display first few rows
print("PROMOTION Dataset:")
print(promotion.head())
print("\nSTORE Dataset:")
print(store.head())
print("\nSUPPORT_TICKET Dataset:")
print(support_ticket.head())

```

```python
import pandas as pd
import numpy as np

# --- 1. Promotion cleaning ---
promotion = promotion.copy()

# Convert date columns to datetime (Fixed: removed infer_datetime_format)
promotion['start_date'] = pd.to_datetime(promotion['start_date'], errors='coerce')
promotion['end_date'] = pd.to_datetime(promotion['end_date'], errors='coerce')

# Ensure discount_value is numeric
promotion['discount_value'] = pd.to_numeric(
    promotion['discount_value'].astype(str).str.replace(r'[^0-9.\-]', '', regex=True),
    errors='coerce'
)

# Handle missing values
promotion['promo_type'] = promotion['promo_type'].astype(str).str.strip().replace({'nan': None})
promotion['promo_type'] = promotion['promo_type'].fillna('Unknown')
promotion['discount_value'] = promotion['discount_value'].fillna(promotion['discount_value'].median())
promotion = promotion.dropna(subset=['promo_id'])

# --- 2. Store cleaning ---
store = store.copy()

# Normalize store city, state and type to Title Case
store['store_city'] = store['store_city'].astype(str).str.strip().str.title()
store['store_state'] = store['store_state'].astype(str).str.strip().str.title()
store['store_type'] = store['store_type'].astype(str).str.strip().str.title()

# Handle missing values
store['store_city'] = store['store_city'].replace({'Nan': None, 'nan': None}).fillna('Unknown')
store['store_state'] = store['store_state'].replace({'Nan': None, 'nan': None}).fillna('Unknown')
store['store_type'] = store['store_type'].replace({'Nan': None, 'nan': None}).fillna('Unknown')
store = store.dropna(subset=['store_id'])

# --- 3. Support ticket cleaning ---
support_ticket = support_ticket.copy()

# Convert created_date to datetime (Fixed: removed infer_datetime_format)
support_ticket['created_date'] = pd.to_datetime(support_ticket['created_date'], errors='coerce')

# Ensure resolution_time_hr is numeric
support_ticket['resolution_time_hr'] = pd.to_numeric(
    support_ticket['resolution_time_hr'],
    errors='coerce'
)

# Derive resolution_time_days
support_ticket['resolution_time_days'] = (support_ticket['resolution_time_hr'] / 24).round(2)

# Handle missing values
support_ticket['status'] = support_ticket['status'].astype(str).str.strip().replace({'nan': None}).fillna('Unknown')
support_ticket['issue_type'] = support_ticket['issue_type'].astype(str).str.strip().replace({'nan': None}).fillna('Unknown')
support_ticket['resolution_time_hr'] = support_ticket['resolution_time_hr'].fillna(
    support_ticket['resolution_time_hr'].median()
)
support_ticket = support_ticket.dropna(subset=['ticket_id'])

# --- 4. Validation checks ---
print('--- PROMOTION Summary ---')
print(promotion.info())
print(promotion.head(3))

print('\n--- STORE Summary ---')
print(store.info())
print(store.head(3))

print('\n--- SUPPORT_TICKET Summary ---')
print(support_ticket.info())
print(support_ticket.head(3))
```

```python
#Save cleaned datasets
promotion.to_csv('../Cleaned_DataSet/PROMOTION.csv', index=False)
store.to_csv('../Cleaned_DataSet/STORE.csv', index=False)
support_ticket.to_csv('../Cleaned_DataSet/SUPPORT_TICKET.csv', index=False)
print("Cleaned datasets saved to Cleaned_DataSet folder")
```
