## PRICE_CHANGE & ORDER_ITEM

```python
import pandas as pd

# Import PRICE_CHANGE and ORDER_ITEM files
price_change = pd.read_csv('../DataSet/PRICE_CHANGE.csv')
order_item = pd.read_csv('../DataSet/ORDER_ITEM.csv')

# Display first few rows
print("PRICE_CHANGE Dataset:")
print(price_change.head())
print("\nORDER_ITEM Dataset:")
print(order_item.head())

```

```python
print(price_change.isnull().sum())
print(order_item.isnull().sum())
```

```python
order_item['promo_id'] = order_item['promo_id'].fillna('No Promo')
```

```python
print(price_change.duplicated().sum())
print(order_item.duplicated().sum())
```
